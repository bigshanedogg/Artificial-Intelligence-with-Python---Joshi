```python
import gym
```

# Reinforcement Learning tutorial...

```python
#input_env = "cartpole"
input_env = "mountaincar"
```

```python
name_map = {"cartpole" : "CartPole-v0",
            "mountaincar" : "MountainCar-v0",
            "pendulum" : "Pendulum-v0", 
            "taxi" : "Taxi-v1", 
            "lake" : "FrozenLake-v0"}
```

```python
env = gym.make(name_map[input_env])
env.reset()
```

```python
for _ in range(1000) : #1000회 반복
    #환경 렌더링
    env.render()
    
    #무작위로 선택한 동작 수행
    env.step(env.action_space.sample())
```

```python
#create environment
env = gym.make(name_map[input_env])
```

```python
for _ in range(20) : 
    observation = env.reset()
    
    for i in range(100) : 
        #reder environment
        env.render()
        
        print(observation)

        #perform action
        action = env.action_space.sample()

        #result(observation/reward/state) of action
        observation, reward, done, info = env.step(action)

        if done : 
            print("Episode finished after {} timesteps".format(i+1))
            break
```
